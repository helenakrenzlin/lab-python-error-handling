````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        # The while loop keeps asking until we get a correct answer
        while not valid_quantity:
            try:
                # Attempt to convert the string input to an integer
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                
                # Logical check: Inventory cannot be below zero
                if quantity < 0:
                    print("Invalid quantity! Please enter a non-negative value.")
                else:
                    inventory[product] = quantity
                    valid_quantity = True # This breaks the while loop
                    
            except ValueError:
                # This catches non-numeric inputs like letters or symbols
                print("Invalid input. Please enter a whole number.")
                
    return inventory
````

````python
def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                # We use float() because prices can have decimals (e.g., 10.99)
                price = float(input(f"Enter the price for {product}: "))
                
                if price < 0:
                    print("Error: Price cannot be negative. Please enter a valid price.")
                else:
                    total_price += price
                    valid_price = True  # Move to the next product in the order
                    
            except ValueError:
                print("Error: Invalid input. Please enter a numeric value for the price.")
                
    return total_price
````

````python
def get_customer_orders(inventory):
    customer_orders = set()
    
    # 1. Handle the total number of orders
    valid_count = False
    while not valid_count:
        try:
            num_orders = int(input("How many items would you like to order? "))
            if num_orders < 0:
                print("Error: Number of orders cannot be negative.")
            else:
                valid_count = True
        except ValueError:
            print("Error: Please enter a whole number for the quantity of orders.")

    # 2. Handle each individual product name
    for i in range(num_orders):
        valid_product = False
        while not valid_product:
            product_name = input(f"Enter the name of product {i+1}: ").strip()
            
            # Check if the product exists in the inventory keys
            if product_name not in inventory:
                print(f"Error: '{product_name}' is not available in our inventory.")
            
            # Check if the product is in stock (quantity > 0)
            elif inventory[product_name] <= 0:
                print(f"Error: '{product_name}' is currently out of stock!")
            
            else:
                customer_orders.add(product_name)
                valid_product = True
                
    return customer_orders
````
